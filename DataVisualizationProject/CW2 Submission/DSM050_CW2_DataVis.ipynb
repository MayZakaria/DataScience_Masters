# Employee Job Satisfaction in the New Era of Post-Pandemic across Different Generations Survey Analysis & Visualization

## 1. Research topic and background

### 1.1 Introduction:

This research aims to analyze the evolving criteria for employee job satisfaction across generations, especially in the post-pandemic era, the influence of work-life balance, work arrangements, and gender diversity within a workplace on job satisfaction. We started by explaining our data sources, external and survey data, followed by data overview and pre-processing. Then the analysis section, we answered our research questions considering the above factors. Our main findings included that across all generations, the common preference to receive feedback and communicate is through in-person meetings and Generation Z's first motivator is 'learning opportunities'. Additionally, a strong agreement on gender diversity's importance but uncertainty about gender pay equality policies.

### 1.2 Background and context:

A survey assessed job satisfaction in the hospitality industry among Gen Y and older employees (baby boomers and Gen X), focusing on delight, voice, loyalty, and turnover intent (Kim, Knutson & Choi, 2015). Employee voice had a lesser impact on Gen Y due to their limited experience, while delight had a stronger effect on them. Older generations exhibited higher loyalty and lower turnover intent. Gen Y turnover rates were attributed to their impatience.

Sanders and Karmowska (2020) interviewed employees and managers to explore the impact of flexible work arrangements on job satisfaction and well-being. Flexibility positively influenced job satisfaction but had mixed effects on well-being, including work intensification and blurred work/home boundaries leading to overwork.

Silaban and Margaretha (2021) found an 8.3% influence of work-life balance on job satisfaction. Mas-Machuca, Berbegal-Mirabent, and Alegre (2016) revealed that with managerial support for work-life initiatives, employees perceive a higher level of work-life balance leading to their satisfaction.

FIELDS and BLUM (1997) discovered higher job satisfaction in gender-balanced groups compared to mostly male or female groups. Workgroup gender composition directly affects job satisfaction alongside co-worker satisfaction, salary levels, and job autonomy in the work context.

  1. To explore the factors influencing job satisfaction across different generations.
  2. To describe the impact of flexible work arrangements on job satisfaction
  3. To explore the influence of work-life balance on job satisfaction
  4. To explore the effect of gender diversity within a workplace on different aspects of job satisfaction such as productivity, pay and benefits

### 1.6 Research questions:

  1. Is there any job satisfaction and preferences across different generations
  2. Do flexible work arrangements (such as WFH, hybrid, and flexible scheduling) impact employee satisfaction?
  3. Does work-life balance influence employee perceptions of overall job satisfaction?
  4. Is there a relationship between job satisfaction and gender diversity within the workplace?

### 1.7 Domain concepts:

**Generations**: There are five generations:

  * Traditional generation: also known as the “Silent Generation”, they were born before 1945 and most of them are retired now (Tolbize, 2008)
  * The Baby Boom generation: were born between 1946 and 1964 (Tolbize, 2008)
  * Generation X: were born between 1965 and 1980 (PEW research center, 2019)
  * Generation Y: also known as “Millennials” or “Echo Boomers”, were born between 1981 and 1996 (PEW research center, 2019)
  * Generation Z: were born between 1997 and 2012 (PEW research center, 2019)


  **Job satisfaction**: Job satisfaction can be the feeling of contentment and pride experienced by individuals who are gratified by or fulfilled in their work and perform it effectively based on some factors (Castaneda and Scanlan, 2014).

## 4. Analysis:

In this section, we aim to answer our research questions through visualizations, including grouped bar charts and heatmaps. We'll maintain the same order as our research questions to provide corresponding answers, starting with the following question:

 ***Is there any job satisfaction and preferences across different generations?***

 We've generated three pivot tables to address this question:

* The first table aggregates feedback and recognition preferences by job satisfaction levels and age groups representing generations.

* The second table aggregates communication methods for work-related discussions by job satisfaction levels within age groups representing generations.

* The third pivot table explores motivation preferences in relation to job satisfaction across different age groups.


We used heatmaps to visualize pivot data, displaying preferences across age groups and job satisfaction levels. The values in the tables represent preference sums for each age-satisfaction combination, and heatmap values show the percentage of preferences within each age group. ### 5.1 Conclusion:
